### Classes, Construtores, Atributos e M√©todos
Teste o c√≥digo abaixo no PyCharm. Voc√™ ver√° que n√£o acontecer√° muita coisa. Pode ficar de boa - em breve voc√™ j√° come√ßar√° a ver os resultados. Olha s√≥:

### M√©todos
At√© agora voc√™ viu um exemplo de classe, atributo e objetos em Python. O que acha de pensarmos em um caso com m√©todos? Vamos imaginar a implementa√ß√£o de uma classe chamada `Circulo` que possui 3 atributos (raio, cor da borda e cor do preenchimento do c√≠rculo) e 4 m√©todos (calcular √°rea, calcular per√≠metro, mudar as cores e redimensionar o raio):

```python
import math

class Circulo:
    def __init__(self, raio_circulo, cor_borda, cor_preenchimento):
        self.raio = raio_circulo
        self.cor_borda = cor_borda
        self.cor_preenchimento = cor_preenchimento

    def calcular_area(self):
        return math.pi * self.raio ** 2

    def calcular_perimetro(self):
        return 2 * math.pi * self.raio

    def mudar_cor(self, nova_cor_borda, nova_cor_preenchimento):
        self.cor_borda = nova_cor_borda
        self.cor_preenchimento = nova_cor_preenchimento

    def redimensionar_raio(self, novo_raio):
        self.raio = novo_raio
```

Primeiro: aquele `import math` n√£o tem nenhuma rela√ß√£o com POO. Usamos ele somente para que possamos pegar o valor de $\pi$, e este valor est√° dentro do m√≥dulo `math` do Python. Isso √© bem mais assertivo do que digitar na m√£o $3.1415...$. Aqui, a classe `Circulo` recebe tr√™s atributos em seu construtor: o `raio`, a `cor_borda` e a `cor_preenchimento`.

> Veja que o `self` sempre √© o primeiro par√¢metro. Isso permite que o m√©todo acesse e modifique os atributos do pr√≥prio objeto. O `self` tamb√©m √© usado para diferenciar as vari√°veis locais da classe e os atributos do objeto.

Os m√©todos `calcular_area()` e `calcular_perimetro()` s√£o respons√°veis por calcular a √°rea e o per√≠metro do c√≠rculo, respectivamente. O m√©todo `mudar_cor()` permite que o usu√°rio mude as cores da borda e do preenchimento se quiser. Por fim, o m√©todo `redimensionar_raio()` permite que o usu√°rio altere o raio do c√≠rculo para um novo valor (`novo_raio`).

Vamos criar 3 objetos de `Circulo`? Olha s√≥:

```python
circulo1 = Circulo(3, "amarelo", "azul")
circulo2 = Circulo(5, "verde", "verde")
circulo3 = Circulo(2, "azul", "roxo")
```

Criamos 3 c√≠rculos diferentes, e cada um deles com um raio e cores diferentes, certo? Agora podemos chamar os m√©todos da classe para realizar c√°lculos ou alterar os atributos dos objetos:

```python
print(f'O raio do c√≠rculo 3 √© {circulo3.raio}')
print(f'A √°rea do c√≠rculo 3 √© {circulo3.calcular_area()}')

circulo3.redimensionar_raio(4)
print(f'O novo raio do c√≠rculo 3 √© {circulo3.raio}')
print(f'A nova √°rea do c√≠rculo 3 √© {circulo3.calcular_area()}')

print(circulo1.cor_borda, circulo1.cor_preenchimento)
print(circulo2.calcular_perimetro())
```

Perceba algumas coisas no exemplo acima:
- Acessamos os atributos sempre seguindo o padr√£o `$OBJETO.$ATRIBUTO` e os m√©todos usando `$OBJETO.$M√âTODO($PAR√ÇMETROS)`. Veja que para acessar o `raio` do `circulo3` far√≠amos `circulo3.raio`. Se fosse do `circulo1`, seria `circulo1.raio`.
    - Para chamar os atributos **n√£o usamos** par√™nteses. Para chamar m√©todo usamos par√™nteses. Veja que sempre usamos `circulo1.raio`, e n√£o `circulo1.raio()`. Isto acontece porque o `raio` √© um atributo, e n√£o uma fun√ß√£o (m√©todo).
    - M√©todos podem ou n√£o ter retorno e par√¢metros, igual como aprendemos em fun√ß√µes, anteriormente. No exemplo acima, `redimensionar_raio()` tem um par√¢metro de entrada (que √© o novo raio) e n√£o tem retorno. J√° `calcular_area()` n√£o tem par√¢metro de entrada, mas tem um retorno.

Em nenhum momento informamos o raio para este m√©todo porque ele precisa obrigatoriamente calcular isso usando todos os atributos que j√° est√£o dentro daquele objeto. Se precisarmos alterar por qualquer motivo que seja o raio ou outros atributos precisaremos, obrigatoriamente, chamar outro m√©todo somente para fazer isto.

### Acesso p√∫blico e privado
Os exemplos acima s√≥ tiveram atributos e m√©todos p√∫blicos. E se quis√©ssemos tamb√©m incluir atributos e m√©todos privados? Veja s√≥:

```python
import math

class Circulo:
    def __init__(self, raio_circulo, cor_borda, cor_preenchimento):
        self.__raio = raio_circulo
        self.cor_borda = cor_borda
        self.cor_preenchimento = cor_preenchimento

    def calcular_area(self):
        return math.pi * self.__raio ** 2

    def calcular_perimetro(self):
        return 2 * math.pi * self.__raio

    def mudar_cor(self, nova_cor_borda, nova_cor_preenchimento):
        self.cor_borda = nova_cor_borda
        self.cor_preenchimento = nova_cor_preenchimento

    def redimensionar_raio(self, novo_raio):
        self.__raio = novo_raio

    def __metodo_privado(self):
        print("Este √© um m√©todo privado.")
```

Veja que o atributo raio foi modificado para `__raio` e que o m√©todo privado foi criado com o prefixo de dois underscores (`__metodo_privado`). Isso indica que esses elementos s√£o privados e s√≥ devem ser acessados internamente na classe. Simples assim. Isto n√£o significa que o acesso externo ser√° *proibido*: da mesma forma que as pessoas ainda podem pisar na grama mesmo que tenha uma placa escrito *"Proibido pisar na grama"*, outros m√©todos poder√£o acessar estes atributos e m√©todos privados. Os dois underscores (`__`) servem para *recomendar* que n√£o se acesse externamente, somente. Por isso, n√£o entenda que isto funciona como algum tipo de trava de seguran√ßa neste sentido.

### Heran√ßa e Polimorfismo
Beleza, e para testarmos heran√ßas? Bom, o c√≠rculo n√£o deixa de ser uma forma, n√£o √©? Assim como o ret√¢ngulo tamb√©m √©. Ent√£o, vamos mudar um pouco as coisas. Teste o c√≥digo abaixo no PyCharm:

```python
import math

class Forma:
    def calcular_area(self):
        return "Ainda n√£o tenho uma l√≥gica implementada para calcular esta √°rea."
    
    def calcular_perimetro(self):
        return "Ainda n√£o tenho uma l√≥gica implementada para calcular este per√≠metro."

class Circulo(Forma):
    def __init__(self, raio):
        self.raio = raio
    
    def calcular_perimetro(self):
        return 2 * math.pi * self.raio
    
    def redimensionar_raio(self, novo_raio):
        self.raio = novo_raio

class Retangulo(Forma):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
    
    def redimensionar_base(self, nova_base):
        self.base = nova_base
        
    def redimensionar_altura(self, nova_altura):
        self.altura = nova_altura
        
circulo1 = Circulo(3)
retangulo1 = Retangulo(10, 2)

print(circulo1.calcular_area())
print(circulo1.calcular_perimetro())

print(retangulo1.calcular_area())
print(retangulo1.calcular_perimetro())
```

#### Heran√ßa
A classe `Forma` √© uma classe base que cont√©m os m√©todos que todas as outras classes ir√£o herdar. A partir dela, foram criadas as subclasses `Circulo` e `Retangulo`, cada uma com seus pr√≥prios m√©todos e atributos. Perceba que temos agora `class Circulo(Forma)` e `class Retangulo(Forma)`. Este `(Forma)` presente em ambos os casos √© o que diz que estas duas classes s√£o herdadas de `Forma`.

Veja tamb√©m que cada classe herdada possui um construtor diferente: o c√≠rculo possui somente um par√¢metro no construtor (`raio`) e o ret√¢ngulo possui dois (`base` e `altura`). Al√©m disso, cada classe herdada pode ter os seus pr√≥prios m√©todos.

A classe `Circulo`, por exemplo, possui o m√©todo `redimensionar_raio()`, que permite alterar o valor do raio do c√≠rculo. A classe `Retangulo` possui a `redimensionar_base()` e `redimensionar_altura()`.

Ao rodar o c√≥digo, o seguinte resultado aparecer√°:

```
Ainda n√£o tenho uma l√≥gica implementada para calcular esta √°rea.
18.84955592153876
Ainda n√£o tenho uma l√≥gica implementada para calcular esta √°rea.
Ainda n√£o tenho uma l√≥gica implementada para calcular este per√≠metro.
```

Veja que somente a segunda linha retornou um valor num√©rico. Se olharmos o c√≥digo, ver√° que a classe `Forma` retorna somente uma mensagem para os m√©todos `calcular_area()` e `calcular_perimetro()`, e somente a classe `Circulo` reimplementou o m√©todo `calcular_perimetro()` para que ela tenha a sua pr√≥pria l√≥gica.

> üëâ Ou seja: n√£o precisamos implementar todos os m√©todos da superclasse para todas as classes herdadas. Se n√£o implementarmos, a classe herdada usar√° a mesma l√≥gica implementada pela superclasse. Como n√£o implementamos a nossa pr√≥pria l√≥gica de `calcular_area()` para `Circulo` e `Retangulo` e nem implementamos a nossa l√≥gica de `calcular_perimetro()` para o `Retangulo`, o algoritmo usou a l√≥gica que existia l√° em `Forma`.

#### Polimorfismo
Ah, e o polimorfismo! Achou que eu ia me esquecer? Vamos alterar o c√≥digo acima para implementar a l√≥gica de `calcular_perimetro()` e `calcular_area()` para as duas classes herdadas:

```python
import math

class Forma:
    def calcular_area(self):
        return "Ainda n√£o tenho uma l√≥gica implementada para calcular esta √°rea."
    
    def calcular_perimetro(self):
        return "Ainda n√£o tenho uma l√≥gica implementada para calcular este per√≠metro."

class Circulo(Forma):
    def __init__(self, raio):
        self.raio = raio
    
    def calcular_area(self):
        return math.pi * self.raio ** 2

    def calcular_perimetro(self):
        return 2 * math.pi * self.raio
    
    def redimensionar_raio(self, novo_raio):
        self.raio = novo_raio

class Retangulo(Forma):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
    
    def calcular_area(self):
        return self.base * self.altura

    def calcular_perimetro(self):
        return 2 * (self.base + self.altura)
    
    def redimensionar_base(self, nova_base):
        self.base = nova_base
        
    def redimensionar_altura(self, nova_altura):
        self.altura = nova_altura
        
circulo1 = Circulo(3)
retangulo1 = Retangulo(10, 2)

print(circulo1.calcular_area())
print(circulo1.calcular_perimetro())

print(retangulo1.calcular_area())
print(retangulo1.calcular_perimetro())
```

Se voc√™ rodar este c√≥digo ver√° que teremos o seguinte resultado:

```
28.274333882308138
18.84955592153876
20
24
```

Consegue ver que tanto para o c√≠rculo quanto para o ret√¢ngulo pedimos pelo mesmo nome do m√©todo? Ou seja: para ambos, chamamos o `calcular_area()` e `calcular_retangulo()`. Por outro lado, os dois possuem l√≥gicas de implementa√ß√£o totalmente diferentes. Isto √© um exemplo pr√°tico de polimorfismo.

Quer outro exemplo de polimorfismo que j√° existe no Python? O operador `+`. Veja que ele consegue somar n√∫meros (`2 + 2` ser√° igual a `4`), mas ele tamb√©m funciona para string (`"Teste" + "Chinelo"` ser√° igual a `TesteChinelo`). Perceba que o mesmo operador tem comportamentos diferentes para tipos de dados diferentes: isto tamb√©m √© polimorfismo.

**2. Crie uma classe `Calculadora` com os m√©todos `somar`, `subtrair`, `multiplicar` e `dividir`. Crie um objeto dessa classe e utilize os m√©todos para realizar opera√ß√µes aritm√©ticas simples.**

```python
class Calculadora:
    def somar(self, a, b):
        return a + b
    
    def subtrair(self, a, b):
        return a - b
    
    def multiplicar(self, a, b):
        return a * b
    
    def dividir(self, a, b):
        if b == 0:
            return "N√£o √© poss√≠vel dividir por zero"
        else:
            return a / b
        
# Cria um objeto da classe Calculadora
minha_calculadora = Calculadora()

# Realiza algumas opera√ß√µes aritm√©ticas simples
print(minha_calculadora.somar(2, 3))  # Imprime 5
print(minha_calculadora.subtrair(5, 2))  # Imprime 3
print(minha_calculadora.multiplicar(4, 6))  # Imprime 24
print(minha_calculadora.dividir(10, 5))  # Imprime 2.0
print(minha_calculadora.dividir(10, 0))  # Imprime "N√£o √© poss√≠vel dividir por zero"
```

Percebeu que este c√≥digo n√£o tem o construtor (`def __init__`)? O c√≥digo n√£o tem construtor porque ele n√£o precisa de nenhum atributo espec√≠fico para funcionar corretamente. Todas as opera√ß√µes aritm√©ticas podem ser realizadas apenas com os par√¢metros de entrada dos m√©todos. Por isso, n√£o √© necess√°rio criar um construtor que inicialize algum atributo da classe.

Veja que este √© um caso espec√≠fico em que n√£o precisamos guardar nenhum atributo antes para calcular. Como uma calculadora √© usada para testar diferentes combina√ß√µes de valores, o prop√≥sito dela realmente n√£o precisa de nenhum construtor neste cen√°rio.

**3. Crie uma classe `Funcionario` com os atributos `nome` e `salario`. Em seguida, crie um m√©todo para calcular o sal√°rio anual do funcion√°rio, incluindo o d√©cimo-terceiro sal√°rio. Crie um objeto dessa classe com sal√°rio definido e exiba na tela o sal√°rio anual.**

```python
class Funcionario:
    def __init__(self, nome, salario):
        self.nome = nome
        self.salario = salario
    
    def calcular_salario_anual(self):
        return self.salario * 13
    
# Cria um objeto da classe Funcionario
funcionario1 = Funcionario("Jo√£o", 3000)

# Calcula o sal√°rio anual do funcion√°rio
salario_anual = funcionario1.calcular_salario_anual()

# Exibe na tela o sal√°rio anual do funcion√°rio
print("O sal√°rio anual do funcion√°rio {} √© R$ {:.2f}".format(funcionario1.nome, salario_anual))
```
